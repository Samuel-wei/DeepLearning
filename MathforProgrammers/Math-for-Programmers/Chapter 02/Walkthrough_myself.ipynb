```python
def distance(v1,v2):
    return length(subtract(v1,v2))
```

```python
def perimeter(vectors):
    distances = [distance(vectors[i],vector[(i+1)%len(vectors)])
                 for i in range(0, len(vectors))]
    return sum(distances)
```
